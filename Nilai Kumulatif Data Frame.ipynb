```python
import pandas as pd
print(pd.__version__)
```

```python
d = {'driver':['verstappen', 'hamilton', 'leclerc', 'sainz', 'perez', 'Ricciardo', 'Vettel', 'Norris'], 
     'podium':[1, 2, 3, 4, 5, 6, 7, 8]}
df = pd.DataFrame(d)
df
```

```python
df['podium'].cumsum().to_frame()
```

```python
df['jumlah_podium_kumulatif'] = df['podium'].cumsum()
df
```

```python
df['jumlah_podium_kumulatif_tiap_driver'] = df.groupby('driver')['podium'].cumsum()
df
```

```python
df['cummax'] = df['podium'].cummax()
df
```

```python
df['cummin'] = df['podium'].cummin()
df
```

```python
df['cumprod'] = df['podium'].cumprod()
df
```
